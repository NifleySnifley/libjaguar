```python
import numpy as np
import pickle
from matplotlib import pyplot as plt
import scipy as sp
```

```python
(voltage, rps) = pickle.load(open("./profile.pickle", 'rb'))
```

```python
plt.plot(rps)
```

```python
SAMPLE_RATE = 1.0/(0.01)

# plt.plot(sp.signal.decimate(rps, 6))

O = 10 # Order
# All over 2hz... lol
F = sp.signal.butter(O, 2, fs=SAMPLE_RATE, btype='low', analog=False)
rps_filt = sp.signal.lfilter(*F,rps)
plt.plot(rps_filt)
```

```python
m,b = np.polyfit(voltage, rps_filt, 1)
plt.plot(voltage, rps_filt)
plt.plot(voltage, b+voltage*m)
xint = (-b/m)
print(xint)
```
